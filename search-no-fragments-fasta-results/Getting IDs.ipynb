```python
import pandas as pd
```

```python
file = 'no_fragments-Copy1.fasta'
```

```python
d = {}

with open(file) as f:
    for line in f:
        if len(line) > 1:
            if '>' in line:
                key = line.strip()
                d[key] = ""
            else:
                d[key] += line.strip() + "+"

d = {key: d[key][:-1] for key in d}
print(d)
```

```python
seq = pd.DataFrame.from_dict(d, orient ='index')
```

```python
seq
```

```python
seq = seq.reset_index()
```

```python
seq
```

```python
seq = seq.replace('>tr|', '', regex=True).astype(str)
```

```python
seq[['split1','split2']] = seq['index'].str.split('|', 1, expand=True)
```

```python
seq
```

```python
seq[['split3','split4']] = seq['split2'].str.split('|', 1, expand=True)
```

```python
seq
```

```python
IDs = pd.DataFrame(seq['split3'])
```

```python
IDs = IDs.set_index('split3')
```
